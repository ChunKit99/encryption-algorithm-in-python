```python

"""
Process of encrypt step by step
"""
def is_prime(num):
  flag = True
  if num > 1:
      for i in range(2, num):
          if (num % i) == 0:
              flag = False
              break          
  return flag


def calc_d(e, r):
  i = 1
  while ((i * e) % r ) != 1:
    i = i + 1
  return i
    
def gcd_of(x, y):
  while y > 0:
    t = y
    y = x % y;
    x = t
  return x

def get_all_e(num):
  result_e = []
  for i in range(2, num):
    if gcd_of(num, i) == 1:
      result_e.append(i)
  return result_e

def calcualte_cipher(plaintext, d, n):
  result = 1
  for i in range(d):
    result = (result * plaintext) % n
  return result

def alphaToNumList(string):
  output = []
  for character in string:
    number = ord(character) - 96 # if a == 1, b == 2
    output.append(number)
  return output

def NumListToString(num):
  character = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q','r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
  res = ""
  for i in range(0, len(num)):
    res = res + character[num[i]-1] # num[i]-1 if a == 1, b == 2
  return res

def NumListToString2(num):
  character = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q','r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
  res = ""
  for i in range(0, len(num)):
    temp = num[i]

    if temp > 26:
      temp = temp % 26
    res = res + character[temp-1] # temp-1 if a == 1, b == 2
  return res

"""
return a list that contain unique element in string
the list is Capital letter
"""
def unique(list1):
    unique_list = []
    for x in list1:
        if x.upper() not in unique_list:
            unique_list.append(x.upper())
    return unique_list

"""
return a list that contain all element in string
the list is Capital letter
"""

def Convert(string):
    list1=[]
    list1[:0]=string.upper()
    return list1

"""
return a list that contain index unique element in string
the list is Capital letter
"""

def indexUnique(plaintext, uniquelist):
  index_list_unique = []
  lsString = Convert(plaintext)
  for i in range(len(uniquelist)):
    index_list_unique.append(lsString.index(uniquelist[i]))
  return  index_list_unique 


print('Step 1. Calculate N which is a product of two distinct prime numbers p and q')

# change the p and q value
p = int(input('p = (enter number or press enter to default)')or'17')
q = int(input('q = (enter number or press enter to default)')or'31')

if not(is_prime(p) and is_prime(q) and q != p):
  print("Warning!! check if p and q are distinct and prime numbers")
else:
  n = p * q
  print('')
  print(f'N = {p} * {q} = {n}')

  print('')
  print('Step 2. Find θ(N) which is (p-1) * (q-1)')
  r = (p-1) * (q-1)
  print('')
  print(f'θ(N) = ({p}-1) * ({q}-1) = {r}')

  print('')
  print('Step 3. Select e such that gcd(θ(N),e) = 1 and 1 < e < θ(N)')
  possible_e = get_all_e(r)

  print("Possible Values of e: ")
  print(possible_e)
  print('')
  index = int(input('Enter e: '))
  eselect = possible_e[possible_e.index(index)]
  if not (eselect in possible_e):
    print(f"{eselect} not exist in list")
  else:
    print(f'Selected e is {eselect}')
    print('')
    print('Step 4. Calculate d such that d*e mod(θ(N)) = 1')
    d = calc_d(eselect, r)
    print(f'd = {d}')

    print('')
    print('Step 5. Set public key and private key')
    print(f'Public Key (e , n) = ({eselect} , {n})')
    print(f'Private Key (d , n) = ({d} , {n})')
    print('')
    print('in this case, the ciphertext will encrypt by using private key')
    print('')
    print('Step 6. Enter plaintext message M to encrypt such that M < N ( C = M d (mod n) )')
    print('')

    # no spacebar for plaintext
    print('no spacebar for plaintext')
    plaintext = (input('Enter text to encrypt: ').lower())
    plainnumberlist = alphaToNumList(plaintext)

    plainnumber = NumListToString(plainnumberlist)
    Clist = []

    for plain in plainnumberlist:
      if plain > n:
       print(f'Error !! the integer number less than {n}')
      else:
        Clist.append(calcualte_cipher(plain, e, n)) # using e(public) to encrypt

    C = NumListToString2(Clist)
    
    print('')
    print(f'Plaintext message M = {plaintext.upper()}')
    print(f'Ciphertext message C = {C.upper()}')
    print(' ')
    print(f'Plaintext message M in number= {plainnumberlist}')
    print(f'Ciphertext message C in number= {Clist}')
    print(' ')

    uniquestringlist = unique(plaintext)
    print('Unique element')
    print(uniquestringlist)
    index_list_unique = indexUnique(plaintext, uniquestringlist)
    print('Unique element index')
    print(index_list_unique)

    print(' ')
    print("Calculation of encryption")
    print(' ')
    for index in index_list_unique:
      i = index
      temp  = Clist[i]
      if temp > 26:
        temp = temp % 26
        print(f'{plaintext[i].upper()}: C = M^e mod n = {plainnumberlist[i]}^{e} mod {n} = {Clist[i]}; {Clist[i]} mod {26} = {temp} = {C[i].upper()}')
      else:
        print(f'{plaintext[i].upper()}: C = M^e mod n = {plainnumberlist[i]}^{e} mod {n} = {Clist[i]} = {C[i].upper()}')
```
