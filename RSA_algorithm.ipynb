```python
def is_prime(num):
  flag = True
  if num > 1:
    for i in range(2, num):
      if (num % i) == 0:
        flag = False
        break          
  return flag

def calc_d(e, r):
  i = 1
  while ((i * e) % r ) != 1:
    i = i + 1
  return i
    
def gcd_of(x, y):
  while y > 0:
    t = y
    y = x % y;
    x = t
  return x

def get_e(num):
  result_e = []
  for i in range(2, num):
    if gcd_of(num, i) == 1:
      result_e.append(i)
  return result_e[-2]

def convert_to_number(String):
  output = []
  for i in range(len(String)):
    char = String[i]
    if (char.isupper()):
      # A = 65
      num = ord(char) - 64
    else:
      # a = 97
      num = ord(char) - 96
    output.append(num)
  return output

def convert_to_string(number):
  output = ''
  for i in range(len(number)):
    num = number[i]
    if num > 26:
      num = num % 26  
    char = chr(num+64) 
    output += char
  return output

def calculate(text, exponent, modulus):
  output = [] # number list
  for i in range(len(text)):
    num = text[i]
    c = (num ** exponent) % modulus
    output.append(c)
  return output

def main():

  # p and q is prime number

  p = 739
  q = 829

  if not(is_prime(p) and is_prime(q) and q != p):
    print('warning! Check p and q are distinct and prime numbers')
    return
  
  print('')
  print(f'p = {p}, q = {q}')

  n = p * q

  print('')
  print(f'N = {p} * {q} = {n}')

  r = (p-1) * (q-1)

  print('')
  print(f'θ(N) = ({p}-1) * ({q}-1) = {r}')

  
  e = get_e(r)
  print('')
  print(f'e = {e}')
  
  d = calc_d(e, r)

  print('')
  print(f'd = {d}')

  print('')

  print(f'Public Key (e , n) = ({e} , {n})')
  print(f'Private Key (d , n) = ({d} , {n})')
  print('')

  plaintext = 'HELLOHOWAREYOU'
  plaintext_number_list = convert_to_number(plaintext)
  ciphertext_number_list = calculate(plaintext_number_list, d, n)

  ciphertext = convert_to_string(ciphertext_number_list)

  print('')
  print(f'Plaintext message M = {plaintext}')
  print(f'Plaintext message M in number = {plaintext_number_list}')
  print(f'Ciphertext message C = {ciphertext}')
  print(f'Ciphertext message C in number = {ciphertext_number_list}')

  # process decrypt
  # rule using the nymber list of Ciphertext
  # output using upper case
  decrypt_text_number_list = calculate(ciphertext_number_list, e, n)
  decrypt_text = convert_to_string(decrypt_text_number_list)

  print(f'Decrypt message M = {decrypt_text}')

if __name__ == "__main__":
  main()
```
